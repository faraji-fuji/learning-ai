## Matplotlib Example Workflow

```python
# 0. Import matplot lib and get it ready for plotting in jupyter

%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup Plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot", xlabel='x-axis', ylabel='y-axis')

# 5. Save & Show (the whole figure)
fig.savefig("figure.png") 
```

```python
# Create some data
x = np.linspace(1, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x,  x**2);
```

```python
# Make plot from a dictionary
butter_prices = {
    "almond":100,
    "peanut":85,
    "cashew":120
}
    
fig, ax = plt.subplots()
ax.bar(butter_prices.keys(), butter_prices.values());
ax.set(title="Nut Butter Prices", xlabel="Nut Butter", ylabel="Price (KES)")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()))
plt.show()
```

```python
#  Make some data for histograms
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
```

### Two options for subplots 

```python
# Subplot option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each different axis
ax1.plot(x , x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(butter_prices.keys(), butter_prices.values())
ax4.hist(np.random.randn(1000))

# Plot to each different axis
plt.show()
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(butter_prices.keys(), butter_prices.values())
ax[1,1].hist(np.random.randn(1000))
plt.show()
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# Create a histogram
heart_disease['age'].plot.hist(bins=10);
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize =(10, 30), subplots=True);
```

```python
over_50 = heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter", x='age', y='chol', c='target');
```

```python
# OO method with pyplot
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x='age', y='chol', c='target', ax=ax);

# ax.set_xlim([45, 100])
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# plot the data
scatter = ax.scatter(x=over_50['age'], 
                     y=over_50['chol'],
                     c=over_50['target'])

# Customize
ax.set(title="Heart Disiease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal axis
ax.axhline(y=over_50['chol'].mean(),
           linestyle='--')

plt.show()
```
